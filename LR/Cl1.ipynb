```python
import pandas as pd
import numpy as np
from sklearn.cluster import AgglomerativeClustering
```

```python
df = pd.read_csv("cities_r2.csv")
```

```python
df
```

```python
x=df[['sex_ratio','effective_literacy_rate_total']]
x
```

```python
cl=[]

from sklearn.cluster import KMeans
for k in range(1, 18) :
    km=KMeans(n_clusters=k).fit(x)
    cl.append(km.inertia_)
cl
    
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1, 18), cl, 'r-')
plt.grid()
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silh = []

for k in range(2, 18):
    ss=KMeans(n_clusters=k,n_init=100)
    labels=ss.fit_predict(x)
    silh.append(silhouette_score(x, labels))
silh
```

```python
plt.plot(range(2,18),silh,'r-')
plt.show()
```

```python
km=AgglomerativeClustering(n_clusters=2)
labels=km.fit_predict(x)

labels
```

```python
import seaborn as sns
sns.scatterplot(x = df['effective_literacy_rate_total'], y = df['total_graduates'], c = labels)
plt.show()
```
